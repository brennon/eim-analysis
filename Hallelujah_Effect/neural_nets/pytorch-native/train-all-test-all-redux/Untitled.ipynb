```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
y_hats = np.linspace(0., 1.0, 100)
```

```python
y_ones = np.ones(y_hats.shape)
y_zeros = np.zeros(y_hats.shape)
```

```python
def ce(y, y_hat):
    eps = 1e-24
    return -y * np.log(y_hat + eps) - (1. - y) * np.log(1. - y_hat + eps)
```

```python
loss_ones = [ce(1., y_hat) for y_hat in y_hats]
loss_zeros = [ce(0., y_hat) for y_hat in y_hats]
```

```python
plt.plot(y_hats, loss_ones)
plt.plot(y_hats, loss_zeros)
plt.xlim([0.01, 0.99])
plt.ylim([0, 5])
plt.yticks(np.arange(0,5.5,.5))
plt.tight_layout()
```
